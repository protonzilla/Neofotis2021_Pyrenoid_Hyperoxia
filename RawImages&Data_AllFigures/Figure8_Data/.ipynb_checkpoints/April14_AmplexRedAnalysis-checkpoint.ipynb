```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
from scipy.stats import ks_2samp
idx = pd.IndexSlice
```

```python

# just for standard curve
def get_585_vals(df):
    df.index = df.iloc[:,0]
    return df.loc[585,df.columns[1::2]]

def get_fullspec_vals(df):
    df.index = df.iloc[:,0]
    return df.iloc[3:,1::2]
```

```python
# standard curve for h202 
standard_curve_data = get_585_vals(pd.read_excel ("PeterNickStandardCurve1_WithLabels.xls"))
x = np.array([float(s[:-2]) for s in standard_curve_data.keys()])
y = standard_curve_data.values.astype(float)


hpr_m,hpr_b = np.polyfit(x, y, 1)

plt.figure()
plt.scatter (x,y)
plt.plot( x, hpr_m*x + hpr_b )
plt.show () 

def compute_hpr( data_585 ):
    return ( data_585 - hpr_b ) / hpr_m

# x-units uM
# y-units spec
```

```python
# standard curve for protein
standard_curve_data = pd.read_excel ("BradfordAssayCurve.xlsx", skiprows=2)

x_label = "A595"
y_label = "Protein Amount "


x = standard_curve_data[x_label]
y = standard_curve_data[y_label]


protein_m,protein_b = np.polyfit(x, y, 1)

plt.figure()
plt.scatter (x,y)
plt.plot( x, protein_m*x + protein_b )
plt.xlabel(x_label)
plt.ylabel(y_label)
plt.show () 

def compute_protein( data ):
    return ( data - protein_b ) / protein_m

# 
```

```python

# colors for plots
all_strain_colors = {
    "c2343":["darkgreen","limegreen","lime"],
    "c1009": ["darkblue","dodgerblue","aqua"]
}

def get_fullspec_vals_mar19(df):

    # data with incomplete column labels
    result = df.iloc[5:,1::3]

    # get the other parts of the column labels
    column_ab_labels = df.iloc[0,0::3].values
    column_dates = [str(d)[5:10] for d in df.iloc[0,1::3].values]
    column_strains = df.columns[0::3]

    # assign new 4-part column labels
    result.columns = pd.MultiIndex.from_tuples(
        zip( result.columns, column_ab_labels, column_dates, column_strains ), 
        names=["time","ab_label","date","strain"])
    
    # assign row labels
    result.index = df.iloc[:,0][5:].values    
    
    return result

# normalize data and estimate hpr based on standard curve
# input one or more filenames (excel workbooks)
def do_analysis_mar19(filenames, 
                      norm_label=None, h2o2_ylabel=None,
                      all_strains=["c2343","c1009"], 
                      all_hours=["0 Hours","6 Hours","31 Hours"]):
    
    
    all_norm_data = {}
    all_hpr_estimates = {}
    
    # from each data file, get normalized fullspec values and h202 values
    for filename in filenames:
        df = get_fullspec_vals_mar19(pd.read_excel(filename))

        # plot raw data
#         plt.figure()
#         for label in df.columns:
#             plt.plot( df[label] )
#         #plt.legend()
#         plt.title( "raw" )
#         plt.show()

        # subtract blank from all columns
        blank_data = df.loc[:, idx["Blank",:,:,:]]
        if len(blank_data.columns) != 1:
            raise Exception( "Should have exactly one column with label \"Blank\"" )
        df_norm = df.subtract( blank_data.iloc[:,0], axis=0 )
        
        # estimate h2o2
        hpr_norm = compute_hpr(df_norm.loc[585])
        df_hpr_norm = pd.DataFrame(columns=hpr_norm.index)
        df_hpr_norm.loc[0] = hpr_norm

        # if necessary, normalize using some data from the same file
        if norm_label is not None:
            norm_values = pd.read_excel(filename, index_col=0).loc[norm_label].values[::3]
            norm_values[norm_values == 0] = np.nan
            
            # special case, estimate protein level based on standard curve
            if norm_label == 'Protein in Extract ':
                norm_values = compute_protein( norm_values )
                
                # print out protein data
                print("protein values used for normalization:")
                for i in range(len(df.columns)):
                    print( df.columns[i], round(norm_values[i],2) )
                print("")
            
            # normalize both the fullspec and the h2o2 values
            df_norm = df_norm / norm_values
            df_hpr_norm = df_hpr_norm / norm_values
            
            # special case, normalize again based on 0-hour data
            if norm_label == 'Protein in Extract ':
                df_2norm = df_norm.copy()
                df_hpr_2norm = df_hpr_norm.copy()
                print( "normalized h2o2 values used for second normalization:")
                
                
                # for each column apply the second normalixzation
                for col in df_hpr_norm.columns[1:]:
                    
                    # if this is a zero-hour column, also print the once-normalized value
                    # it will normalize to itself for the second normalization
                    # it will also be used to normalize other columns for the second normalization
                    if col[0].startswith("0 Hours"):
                        print( col, round(df_hpr_norm[col].values[0],3) )
                    
                    # find corresponding 0-hour column
                    # some parts of the column labels match perfectly,
                    # others match based on prefix only
                    norm_col = next(tcol for tcol in df_hpr_norm.columns if 
                                    tcol[0].startswith("0 Hours") and
                                    tcol[-1].split(".")[0] == col[-1].split(".")[0] and 
                                    tcol[1:-1] == col[1:-1] )
                    
                    # normalize both the fullspec and the h2o2 values
                    df_2norm[col] = df_norm[col] / df_norm[norm_col]
                    df_hpr_2norm[col] = df_hpr_norm[col] / df_hpr_norm[norm_col]
                
                df_norm = df_2norm
                df_hpr_norm = df_hpr_2norm
                    
                print("")
        
        # remove blank data
        df_norm = df_norm.drop( blank_data.columns, axis=1 )
        df_hpr_norm = df_hpr_norm.drop( blank_data.columns, axis=1 )
        
        # keep track of normalized data
        all_norm_data[filename] = df_norm
        all_hpr_estimates[filename] = df_hpr_norm

    # get data for plots, arranged by ab_label and by strain
    # plots are divided differently than the data files
    all_subplot_data_by_strain = {}
    for filename,df_norm in all_norm_data.items():
        all_ab_labels = sorted(list(set(df_norm.columns.get_level_values("ab_label"))))
        for subplot_index,ab_label in enumerate(all_ab_labels):
            df_sub = df_norm.loc[:, idx[:,ab_label,:,:]]
            hpr_sub = all_hpr_estimates[filename].loc[:, idx[:,ab_label,:,:]]
            strain = df_sub.columns.get_level_values("strain")[0].split(".")[0]
            if strain not in all_subplot_data_by_strain.keys():
                all_subplot_data_by_strain[strain] = []
            all_subplot_data_by_strain[strain].append([df_sub,hpr_sub])
            
            
    # show bar graphs with normalized fullspec traces, one subplot per strain
    fig,axes = plt.subplots( 1, len(all_strains), figsize=(15,7), sharey="all")
    for subplot_index,strain in enumerate(all_strains):
        strain_data = all_subplot_data_by_strain[strain]
        strain_colors = all_strain_colors[strain]
        ax = axes[subplot_index]
        
        # for each timepoint
        for color_index,hours in enumerate(all_hours):
            
            # collect data
            values_to_avg = []
            for dataframes in strain_data:
                df_sub = dataframes[0]
                for label in df_sub.columns:
                    if label[0].startswith(hours):
                        values_to_avg.append(df_sub[label])
                    
            # plot one line
            values_to_avg = np.array(values_to_avg, dtype=float)
            y = np.mean(values_to_avg, axis=0)
            yerr = np.std(values_to_avg, axis=0)
            ax.errorbar( range(len(y)), y, yerr, label=hours, color=strain_colors[color_index], capsize=5)
            
        ax.legend()
        ax.set_title( strain )
        ax.set_ylabel( "uM H2O2 normalized by {0}".format( norm_label ))

    # show bar graphs with normalized h2o2, one subplot per strain
    fig,axes = plt.subplots( 1, len(all_strains), figsize=(15,7), sharey="all")
    for subplot_index,strain in enumerate(all_strains):
        strain_data = all_subplot_data_by_strain[strain]
        strain_colors = all_strain_colors[strain]
        ax = axes[subplot_index]
        
        # for each timepoint
        values_by_time = {}
        for x,hours in enumerate(all_hours):
            
            # collect data
            values_to_avg = []
            for dataframes in strain_data:
                df_sub = dataframes[1]
                for label in df_sub.columns:
                    if label[0].startswith(hours):
                        values_to_avg.append(df_sub[label])
            values_by_time[hours] = np.array(values_to_avg).flatten()
                    
            # plot one bar
            values_to_avg = np.array(values_to_avg)
            ax.bar( x, values_to_avg.mean(), yerr = values_to_avg.std(), color=strain_colors[x] )
        
    
        # compute p-values between the sets of h2o2 values
        for i,hours_a in enumerate( all_hours ):
            for hours_b in all_hours[i+1:]:
                _,p_val = ks_2samp(values_by_time[hours_a],values_by_time[hours_b])
                print( "{0} p-value between {1} and {2}: {3}".format( strain, hours_a, hours_b, p_val ) )
            
        ax.set_xticks( range(len(all_hours)))
        ax.set_xticklabels( all_hours )
        ax.set_title( strain )
        if h2o2_ylabel is None:
            h2o2_ylabel = norm_label
        ax.set_ylabel( "uM H2O2 normalized by {0}".format( h2o2_ylabel ))
    fig.suptitle( "Normalized H2O2: {0}".format( norm_label ))
    fig.show()
    
    
```

```python
# all data files are loaded and and analyzed as one set
filenames = [
    "c1009&c2343_SecondTest_FeelsGood_ForNormalization2.xls",
    "c2343&c1009_ThirdTest_FeelsGood_ForNormalization2.xls"
]

#all_norm_labels = [None, 'Total Chlorophyll (Âµg/mL)', 'Average AFDW (mg/L)', 'Protein in Extract ']
```
