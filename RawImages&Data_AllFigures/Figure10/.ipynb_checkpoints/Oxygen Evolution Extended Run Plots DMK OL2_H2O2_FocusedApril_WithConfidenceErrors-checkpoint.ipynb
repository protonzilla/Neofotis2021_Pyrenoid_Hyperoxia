```python
#%matplotlib notebook

import pandas as pd
import matplotlib.pyplot as plt
import numpy as np
from scipy.stats import linregress
```

```python
folder = "/Users/pneofotis/Desktop/DataFromBensSpec/"
#folder = "DataFromBensSpec/"

# start with 4 phi2 time and add more until we have 9
#if I change the number of loops I just have to add the points 

phi2_times=[0, 189.19810000000001, 377.2325, 565.69849999999997, 
            754.02650000000006, 942.094, 1130.44, 1318.86, 1507.42]
while( len(phi2_times) < 18 ):
    phi2_times.append( phi2_times[-1] + (phi2_times[1]-phi2_times[0]))
resp_times = [1832, 1940.2]

# load metadata
master_dict = { "data-by-sample": pd.read_excel(
    folder + "CompiledMetaData_PN10.xlsx", index_col=0) }

#This means just load the data from the flask experiments 
df = master_dict["data-by-sample"]
master_dict["data-by-sample"] = df[df["FromFlask"] == "Y"]
master_dict["data-by-sample"]
```

```python


# remove some rows from the data
df = master_dict["data-by-sample"]
#df = df.drop("C1009_WithEthanol_dee")

#Here is where we decide what strains to exclude 

#remove_strains = ["c4425","c4666"]
remove_strains = ["",""]

for strain in remove_strains:
    strain_rows = df[df["strain"] == strain].index
    print( "removing {0} rows with strain {1}".format( len(strain_rows), strain ))
    df = df.drop(strain_rows)
master_dict["data-by-sample"] = df


master_dict["data-by-sample"]["label"] = master_dict["data-by-sample"].index


# load all raw data files
phi2_dataframes = []
o2_dataframes = []
pH_dataframes = []
Dirk_dataframes = []
for run in master_dict["data-by-sample"].iterrows():
    run = run[1]
    
    raw_phi2_file= folder + run["Spec data"]
    phi2_df = pd.read_table(raw_phi2_file, header=None, names=["Time", "F", "XXX", "YYY"])
    phi2_df["Time"] = phi2_df["Time"] - phi2_df["Time"].values[0]
    
    #Here is what I am adding 
    raw_Dirk_File= folder +run["Dirk Data"]
    Dirk_df = pd.read_table(raw_Dirk_File, header=None, names=["Time", "AAA", "BBB", "ECS"], index_col=False)
       
    adj_o2_file = folder + run["oxygen data"]
    adj_pH_file = folder + run["pH data"]
    oxy_df = pd.read_table(adj_o2_file, sep=",")
    if run["UseIndex"]=="Y":  
        oxy_df["Time"] = oxy_df.index
    pH_df = pd.read_table(adj_pH_file, sep=",")
    
    phi2_dataframes.append( phi2_df )
    o2_dataframes.append( oxy_df )
    pH_dataframes.append( pH_df )
    Dirk_dataframes.append( Dirk_df )
    
master_dict["data-by-sample"]["phi2_df"] = phi2_dataframes
master_dict["data-by-sample"]["o2_df"] = o2_dataframes
master_dict["data-by-sample"]["pH_df"] = pH_dataframes
master_dict["data-by-sample"]["Dirk_df"] = Dirk_dataframes

master_dict["calibration"] = {}
master_dict["calibration"]["raw"] = pd.read_csv(folder + "/GasTest.csv")
```

```python
# check if there are any duplicate row labels
if len(set(master_dict["data-by-sample"].index)) < len(master_dict["data-by-sample"].index):
    raise Exception( "there are duplicate row labels in sample data")
```

```python
# get some statistics from the calibration data

from scipy import stats
slope, intercept, r_value, p_value, std_err = \
    stats.linregress(master_dict["calibration"]["raw"]["Measured"], master_dict["calibration"]["raw"]["Actual"])
master_dict["calibration"]["computed"] = pd.DataFrame.from_dict(
    {"slope":[slope], "intercept":[intercept], "r_value":[r_value], "p_value":[p_value], "std_err":[std_err]},)
    
```

```python
# compute photoparams-over-time for each sample

photoparam_dfs = []
for row in master_dict["data-by-sample"].index:
    number_of_flashes = master_dict["data-by-sample"].loc[row,"number of flashes"]
    result_df = pd.DataFrame( columns=phi2_times[:number_of_flashes] )
    phi2_df = master_dict["data-by-sample"].loc[row,"phi2_df"]
    
    fmpp = phi2_df["F"].max()
    if fmpp > 1:
        print( "found high fmpp value ({0}) for sample \"{1}\". This is the maximum F value, which was at time {2}".format( 
                fmpp, row, phi2_df[phi2_df["F"] == fmpp]["Time"] ) )
    
    for time in result_df.columns:
        phi2_df_sub = phi2_df[(phi2_df["Time"] > time-1000) & (phi2_df["Time"] < time+40)]
        result_df.loc["fs", time] = phi2_df_sub["F"].min()
        result_df.loc["fmp", time] = phi2_df_sub["F"].max()
        result_df.loc["fmpp",time] = fmpp
        
    result_df.loc["phi2",:] = \
        (result_df.loc["fmp",:] - result_df.loc["fs",:]) / result_df.loc["fmp",:]
    
    result_df.loc["ETR",:] = \
        result_df.loc["phi2",:] * 750 * .42
    
    result_df.loc["qE",:] = \
        (result_df.loc["fmpp",:] - result_df.loc["fmp",:]) / result_df.loc["fmp",:]
        
    for time in result_df.columns:
        if result_df.loc["qE",time] > 1:
            print( "found high qE=(fmpp-fmp)/fmp value ({0}) at time {1}: fmp was {2} and fmpp {3}, for sample \"{4}\"".format( 
                    result_df.loc["qE",time], time, result_df.loc["fmp",time], result_df.loc["fmpp",time], row))
        
    if result_df.isnull().values.any():
        print( "got some NaN photoparams for sample: " + row)
        
    photoparam_dfs.append( result_df )    
master_dict["data-by-sample"]["photoparams"] = photoparam_dfs
```

```python
# compute o2 slope and peak for each sample

master_dict["data-by-sample"]["slope_o2"] = ""
master_dict["data-by-sample"]["peak_o2"] = ""

#If you want to see the plots just change Falst to True 
show_plots = True

for row in master_dict["data-by-sample"].iterrows():
    run = row[1]
    
    phi2_df = run["phi2_df"]
    oxy_df = run["o2_df"]
    pH_df = run["pH_df"]

    #show_plots = (row[0] in ['c2343_50ugChl_9flashes', 'c2343_50ugChl_9Flashes2','C1009_50ugChl_Medium1'])
    
    if show_plots:
        fig, ax1 = plt.subplots()
        ax1.plot(oxy_df["Time"]-oxy_df["Time"][0], oxy_df[" Oxygen"], label='[O2] (uM)', color='red')
        ax1.plot( phi2_df["Time"], phi2_df["F"] *500, label='fluorescence', color='green')
        ax1.set_xlabel('time (s)')
        # Make the y-axis label, ticks and tick labels match the line color.
        #ax1.set_ylabel('exp', color='b')
        ax1.tick_params('y', colors='r')
        ax1.legend( )

        ax2 = ax1.twinx()
        #ax2.set_ylabel('sin', color='r')
        ax2.plot(pH_df["Time"]-pH_df["Time"][0], pH_df["pH"], label='pH', color='blue')
        ax2.tick_params('y', colors='b')
        ax2.legend(loc="right")

        # add vertical lines
        for t in phi2_times:
             ax2.axvline(x=t, color="pink", linestyle="dashed")

        for t in [run["start resp time"],run["end resp time"]]:
             ax2.axvline(x=t, color="olive", linestyle="dashed") 

        fig.tight_layout()
        ax1.set_ylim(1, 1500)
        plt.xlim(1, 4500)
        plt.title(run[0])
        plt.show()

    label = run["label"]
    strain = run["strain"]
    x1 = run["start resp time"]
    x2 = run["end resp time"]

    # simple way to get slope (rise over run at two points)
    #y1 = oxy_df[oxy_df["Time"] == x1][" Oxygen"].values[0]
    #y2 = oxy_df[oxy_df["Time"] == x2][" Oxygen"].values[0]
    #slope = (y2-y1)/(x2-x1)

    # use polyfit to get oxygen slope
    if x1 > oxy_df["Time"].max():
        raise Exception("Respiration times are too large for " + label)
    sub_oxy_df = oxy_df[(oxy_df["Time"] >= x1) & (oxy_df["Time"] <= x2)]
    x = sub_oxy_df["Time"].values
    y = sub_oxy_df[" Oxygen"].values
    slope = np.polyfit(x, y, 1)[0]
    
    peak_o2 = sub_oxy_df[" Oxygen"].mean() #(y1+y2)/2
    master_dict["data-by-sample"].loc[row[0],"slope_o2"] = slope
    master_dict["data-by-sample"].loc[row[0],"peak_o2"] = peak_o2
    
        
#     if strain not in combined_o2_slopes_labels.keys():
#         combined_o2_slopes_labels[strain] = [[],[],[]]
#     combined_o2_slopes_labels[strain][0].append(slope)
#     combined_o2_slopes_labels[strain][1].append(peak_o2)
#     combined_o2_slopes_labels[strain][2].append(label)
```

```python
from numpy import nan

# compute more oxygen slopes, for ech phi2 time

# we are going to build one new dataframe for every sample
updated_photoparam_dfs = []

light = 1000

for row in master_dict["data-by-sample"].iterrows():
    run = row[1]
    
    number_of_flashes = master_dict["data-by-sample"].loc[row[0],"number of flashes"]
    
    phi2_df = master_dict["data-by-sample"].loc[row[0],"photoparams"]
    for new_row_label in ["phi2*light", "slope1-slope2", "slope1", "slope2"]:
        if new_row_label not in phi2_df.index:
            s = pd.Series(nan, index=phi2_times[:number_of_flashes])
            s.name = new_row_label
            phi2_df = phi2_df.append( s )
    updated_photoparam_dfs.append( phi2_df )
    
    oxy_df = run["o2_df"]
    
    peak_time = oxy_df[oxy_df[" Oxygen"] == oxy_df[" Oxygen"].max()]["Time"].values[0]
    
    df = pd.DataFrame()
    
    result_df = pd.DataFrame()
    
    # iterating through all the phi2 times
    for time in phi2_times[:number_of_flashes]:
        if time > peak_time:
            print( "time was greater than peak time, getting slope anyways")
            #continue
        
        try:
            x1 = time-20
            x2 = time+20

            # get phi2 value at the timepoint
            phi2 = phi2_df.loc["phi2",time]

            # compute the first slope which is positive, at the phi2 time
            sub_oxy_df = oxy_df[(oxy_df["Time"] >= x1) & (oxy_df["Time"] <= x2)]
            x = sub_oxy_df["Time"].values
            y = sub_oxy_df[" Oxygen"].values
            slope1 = np.polyfit(x, y, 1)[0]

            # compute the second slope which is negative, 
            # ...at some timepoint after the peak oxygen 
            # ...when the oxygen value is the same as it was at the phi2 time
            o2_target = y.mean()
            sub_oxy_df = oxy_df[(oxy_df["Time"] > peak_time)]
            slope2_row_index = (sub_oxy_df[" Oxygen"] - o2_target).abs().argsort().values[0]
            slope2_time = sub_oxy_df["Time"].values[slope2_row_index]
            x1 = slope2_time
            x2 = slope2_time+10
            sub_oxy_df = oxy_df[(oxy_df["Time"] >= x1) & (oxy_df["Time"] <= x2)]
            x = sub_oxy_df["Time"].values
            y = sub_oxy_df[" Oxygen"].values
            slope2 = np.polyfit(x, y, 1)[0]
            
            etr = phi2_df.loc["ETR",time]

            phi2_df.loc["phi2*light",time] = phi2*light
            phi2_df.loc["slope1-slope2",time] = slope1-slope2
            phi2_df.loc["slope1",time] = slope1
            phi2_df.loc["slope2",time] = slope2
            phi2_df.loc["ETR/slope1",time] = etr/slope1
        except Exception as e:
            print(e)
            continue
    
master_dict["data-by-sample"]["photoparams"] = updated_photoparam_dfs
```

```python
# define some groups we will use for aggregating samples later
# this will add a new "data-by-light-strain" dataframe to master_dict

master_dict["data-by-light-strain"] = pd.DataFrame(columns=[
    "strain","number of flashes","BiCarbonate","HydrogenPeroxide","samples"])
df = master_dict["data-by-sample"]

for strain in set(df["strain"]):
    for num_flashes in set(df["number of flashes"]):
        for bicarbonate in set(df["BiCarbonate"]):
            for HydrogenPeroxide in set(df["HydrogenPeroxide"]):
                row_label = "{0}, {1} flashes, {2} bicarbonate, {3} HydrogenPeroxide".format( 
                    strain, num_flashes, bicarbonate, HydrogenPeroxide )
                samples = df[
                             (df["strain"]==strain)
                             & (df["number of flashes"]==num_flashes)
                             & (df["BiCarbonate"]==bicarbonate)
                             & (df["HydrogenPeroxide"]==HydrogenPeroxide)]
                                        
                                             
                master_dict["data-by-light-strain"].loc[row_label] = \
                    [strain, num_flashes, bicarbonate, HydrogenPeroxide, samples]
                                
```

```python
# compute averaged photosynthesis parameters for each of the "data-by-light-strain" groups
avg_photoparam_dfs = []
df = master_dict["data-by-light-strain"]
for row in df.index:
    photoparam_dfs = df.loc[row,"samples"]["photoparams"].values
    if( len(photoparam_dfs) == 0 ):
        avg_photoparam_dfs.append( None )
        continue
    avg_df = pd.DataFrame( index=photoparam_dfs[0].index, columns=photoparam_dfs[0].columns)
    for row in avg_df.index:
        for col in avg_df.columns:
            avg_df.loc[row,col] = np.mean([pdf.loc[row,col] for pdf in photoparam_dfs])
    avg_photoparam_dfs.append( avg_df )
        
master_dict["data-by-light-strain"]["photoparams"] = avg_photoparam_dfs
```

```python
# compute averaged (o2 vs time) dataframe for each of the "data-by-light-strain" groups
# insert the new dataframes into master_dict["data-by-light-strain"]["o2-over-time"]

df = master_dict["data-by-light-strain"]
o2_over_time_dfs = []

for row in df.index:
    samples_df = df.loc[row,"samples"]
    if len(samples_df) == 0:
        #print( "found no data for " + row )
        o2_over_time_dfs.append(None)
        continue
    all_oxy_columns = pd.DataFrame()
    for run in samples_df.iterrows():
        run = run[1]
        run_oxy_df = run["o2_df"]
        all_oxy_columns[run["label"]] = run_oxy_df[" Oxygen"]
        times = run_oxy_df["Time"]

    y = all_oxy_columns.mean(axis=1)
    yerr = all_oxy_columns.std(axis=1)
    x = times[0:len(y)]
    y = y[0:len(x)]
    yerr = yerr[0:len(x)]

    o2_over_time_dfs.append(pd.DataFrame.from_dict(
            {"Time":x,"o2":y, "o2-err":yerr}))
    
master_dict["data-by-light-strain"]["o2-over-time"] = o2_over_time_dfs
```

```python
# OLIVER August 21 2018

# compute averaged (dirk vs time) dataframe for each of the "data-by-light-strain" groups
# insert the new dataframes into master_dict["data-by-light-strain"]["o2-over-time"]

df = master_dict["data-by-light-strain"]
dirk_over_time_dfs = []
for row in df.index:
    samples_df = df.loc[row,"samples"]
    if len(samples_df) == 0:
        #print( "found no data for " + row )
        dirk_over_time_dfs.append(None)
        continue
    all_dirk_columns = pd.DataFrame()
    for run in samples_df.iterrows():
        run = run[1]
        run_dirk_df = run["Dirk_df"]
        all_dirk_columns[run["label"]] = run_dirk_df["ECS"]
        times = run_dirk_df["Time"]

    y = all_dirk_columns.mean(axis=1)
    x = times[0:len(y)]
    y = y[0:len(x)]

    dirk_over_time_dfs.append(pd.DataFrame.from_dict(
            {"Time":x,"ECS":y}))
    
master_dict["data-by-light-strain"]["dirk-over-time"] = dirk_over_time_dfs
```

```python
# OT May 14th 2018
# for each sample, compute the followng values at multiple timepoints:
#   [variable name]   [label on plot]
#   dy                'd[O2]/dt uM/s'
#   dy_o2             'O2 concentration'
# results (dataframes) are added in a new "data-by-sample" column "o2_evolution"

o2_evolution_dataframes = []

for row in master_dict["data-by-sample"].iterrows():
    experiment = row[1]
    oxy_df = experiment["o2_df"]
    y=np.array(oxy_df[" Oxygen"])
    x=np.array(oxy_df["Time"])
    dy=[]
    dy_o2=[]
    dif_time=400
    for index in range(dif_time, len(y)):
        dy.append(y[index]-y[index-dif_time])
        dy_o2.append((y[index]+y[index-dif_time])/2)
        
    o2_evolution_dataframes.append( pd.DataFrame.from_dict( 
            {"d[O2]/dt uM/s": dy, "O2 concentration": dy_o2} ) )
    
master_dict["data-by-sample"]["o2_evolution"] = o2_evolution_dataframes
```

```python
# OT May 14th 2018
# for each sample, compute (oxygen, and oxygen slope / phi2) at multiple timepoints
# (NOT the same timepoints as the previous cell)
# results (dataframes) are added in a new "data-by-sample" column "o2_evolution:fmp"

o2_evolution_phi2_dataframes = []

for row in master_dict["data-by-sample"].iterrows():
    experiment = row[1]
    oxygen = []
    oxygen_slope_over_fmp = []
    all_fmp = []
    oxy_df = experiment["o2_df"]
    phi2_df = experiment["phi2_df"]
    
    for t in phi2_times:
        x1 = t
        x2 = t+40
        y1 = oxy_df[oxy_df["Time"] > x1][" Oxygen"].values[0]
        y2 = oxy_df[oxy_df["Time"] > x2][" Oxygen"].values[0]
        
        phi2_df_sub = phi2_df[phi2_df["Time"] > x1]
        if len(phi2_df_sub) == 0:
            break
        fmp = phi2_df_sub["F"].values[0]
        all_fmp.append(fmp)
        oxygen_slope_over_fmp.append( ((y2-y1)/(x2-x1)) / fmp )
        oxygen.append(y1)
        
    o2_evolution_phi2_dataframes.append( pd.DataFrame.from_dict( 
            {"oxygen slope over fmp": oxygen_slope_over_fmp, "oxygen": oxygen, "fmp": all_fmp} ) )
    
master_dict["data-by-sample"]["o2_evolution:fmp"] = o2_evolution_phi2_dataframes
    
```
